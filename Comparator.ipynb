```python
import laspy
import pandas as pd
import classifications
from collections import defaultdict
```

```python
infile = "data/oakland_part2_ac_classified.las"
predicted = laspy.file.File(infile)
truth = "data/oakland_part2_ac.xyz_label_conf"
df = pd.read_csv(truth, 
                 names=['x','y','z','label','conf'],
                 delimiter = ' ',
                 skiprows=63)
```

```python
def encode_class(key):
    return classifications.encode_class[key]
df['simple_class'] = df.label.apply(encode_class)
```

```python
tp = 0
tn = 0
fp = 0
fn = 0
tracker = defaultdict(lambda: 0)
for i in range(len(predicted.x)):
    print("Step: ", i)
    x = predicted.x[i]
    y = predicted.y[i]
    z = predicted.z[i]
    true = df[(df.x == x) & (df.y == y) & (df.z == z)]['label']
    simple = df[(df.x == x) & (df.y == y) & (df.z == z)]['simple_class']
    print(true)
    if len(true) > 2 or len(true) == 0:
        continue
    true = true[i]
    simple = simple[i]
    pred = predicted.classification[i]
    tracker[true] += 1
    if simple == pred:
        tp += 1
    elif ((simple == 2002 and pred != 30) or (simple == 2001 and pred != 60)):
        fn += 1
    elif ((simple != 2002 and pred == 30) or (simple != 2001 and pred == 60)):
        fp += 1
    else:
        tn += 1
    
```
